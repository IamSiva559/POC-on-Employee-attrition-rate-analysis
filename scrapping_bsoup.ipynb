```python
#randomize headers not to get blocked while scraping
#scrape using different browers everytime
#store headers in dictionary to pass in other file
header = {}
def randomUserAgents():
    head = random.choice(user_agents)
    head = head.split(") ")
    #print(head)
    agent = head[0]+')'

    #print(agent)
    accept = head[1]
    #print(accept)
    header["User-Agent"] = agent
    header["Accept"] = accept
    #print(header)
    return header

#randomUserAgents()
```

```python
url = "https://www.glassdoor.co.in/Reviews/Enjoy-Technology-Reviews-E1090707.htm"
#store randomized user agent headers in head variable
head = randomUserAgents()
```

```python
#function which returns whole text content of a page in lxml format
def soup(url,headers):
    #scrape text content of page and return that lxml text content for every page iteration'''
    #instantiate session object
    session = requests.Session()
    #pass url and headers to get response
    req = session.get(url, headers=headers)
    #print('req variable: ', req)
    #parse text content response using .text and it is lxml formartting (for fastness). you can use html_parser
    # as you wish
    bs = BeautifulSoup(req.text, 'lxml')
    #print('bs variable: ', bs)
    return bs
```

```python
start = time.time()
count = 1
for page in pages:
    #call the function and pass each page link and headers
    bs = soup(page,head)
    #employee reviews
    for x in bs.findAll('li',{'class':'empReview cf'}):

    # ## PK
        a.append(count)
        count += 1

    #rating given by user
        try:
            rating = x.find('span',{'class':'rating'}).find('span')
            review.append(rating.attrs['title'])
        except:
            review.append('None')
    #sub rating like workculture, management etc., 5 sub ratings
        try:
            rate = x.find('ul',{'class':'undecorated'})
            subreviews = rate.findAll('li')
            workLife.append(subreviews[0].find('span').attrs['title'])
        except:
            workLife.append('None')
        #Culture
        try:
            culture.append(subreviews[1].find('span').attrs['title'])
        except:
            culture.append('None')
        #diversity and inclusion
        try:
            diversity_inclusion.append(subreviews[2].find('span').attrs['title'])
        except:
            diversity_inclusion.append('None')
        #career opurtunities
        try:
            careerOpp.append(subreviews[3].find('span').attrs['title'])
        except:
            careerOpp.append('None')
        #benifits
        try:
            compBenefits.append(subreviews[4].find('span').attrs['title'])
        except:
            compBenefits.append('None')
        #SrManagament    
        try:
            srManagement.append(subreviews[5].find('span').attrs['title'])
        except:
            srManagement.append('None')              
    #date review has been given
        try:
            date.append(x.find('time',{'class':'date subtle small'}).text.strip())
        except:
            date.append('None')
    # Summary
    #summary title
        try:
            summar = x.find('h2',{'class':'h2 summary strong mb-xsm mt-0'}).text.strip()
            summ.append(summar)
        except:
            summ.append('None')
    ## Pros
    #pros of the company
        try:
            pro.append(x.find('span',{'data-test':'pros'}).text.strip())
        except:
            pro.append('None')
    ## Cons
    #pros of the company
        try:
            con.append(x.find('span',{'data-test':'cons'}).text.strip())
        except:
            con.append('None')
    ## Advice to Management
        try:
            advice.append(x.find('span',{'data-test':'advice-management'}).text.strip())
        except:
            advice.append('None')

    ## Employee Type
    #whether former employee or current employee
        try:
            employee.append(x.find('span',{'class':"authorJobTitle"}).text.split("-")[0].strip())
        except:
            employee.append('None')
    #designation
        try:
            position.append(x.find('span',{'class':"authorJobTitle"}).text.split("-")[1].strip())
        except:
            position.append('None')
    #location worked
        try:
            work_location.append(x.find('span',{'class':"authorLocation"}).text.strip())
        except:
            work_location.append('None')            
    ## Review Link
    #each employee review link
        link.append(url+(x.find('a',{'class':'reviewLink'}).attrs['href']))
```

```python
#store all list information into dataframe and convert into csv file
df=pd.DataFrame(index=a)
df['date']=date
df['employeeType']=employee
df['position']=position
df['work_location'] = work_location
df['summary']=summ
df['pro']=pro
df['con']=con
#df['advice']=advice
df['overallStar']=review
df['workLifeStar']=workLife
df['cultureStar']=culture
df['careerOppStar']=careerOpp
df['comBenefitsStar']=compBenefits
df['srManagementStar']=srManagement
df['reviewLink']=link

csvName = "enjoy_inc_glassdoor_reviews"
df.to_csv('{}.csv'.format(csvName), sep=',')
print('StartTime = {}\nEnd Time = {}'.format(start, time.time() - start))
```
